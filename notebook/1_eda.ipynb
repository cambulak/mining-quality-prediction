```python
# Hücre 1: Kütüphaneleri İmport Et
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# Uyarıları kapatalım
warnings.filterwarnings('ignore')

# Pandas ayarları (tüm sütunları görmek için)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)

# Hücre 2: Veriyi Oku
# Verinin 'data' klasöründe ve isminin 'mining_data.csv' olduğunu varsayıyoruz
# Tarih formatını otomatik algılaması için parse_dates kullanıyoruz
df = pd.read_csv(r'C:\Users\Sedat\PycharmProjects\mining-quality-prediction\data\MiningProcess_Flotation_Plant_Database.csv', decimal=',', parse_dates=['date'], dayfirst=True)

# İlk 5 satıra bakalım
print("Veri Setinin İlk 5 Satırı:")
display(df.head())

# Hücre 3: Veri Hakkında Genel Bilgi
print("\nVeri Seti Bilgileri:")
print(df.info())

print("\nİstatistiksel Özet:")
display(df.describe().T)

# Hücre 4: Hedef Değişken Analizi (% Silica Concentrate)
# Hedefimiz: % Silica Concentrate (Silika ne kadar az ise o kadar iyi)
plt.figure(figsize=(10, 6))
sns.histplot(df['% Silica Concentrate'], bins=50, kde=True, color='blue')
plt.title('Hedef Değişken Dağılımı: % Silica Concentrate')
plt.xlabel('Silika Oranı')
plt.ylabel('Frekans')
plt.show()

# Hücre 5: Korelasyon Matrisi (Hangi değişkenler birbirini etkiliyor?)
# Bu grafik madencilik sürecini anlamak için çok kritik
plt.figure(figsize=(20, 18))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Değişkenler Arası Korelasyon Matrisi')
plt.show()
```

# 📊 EDA Bulguları ve Değişken Analizi

Bu analizde, flotasyon tesisinden alınan 737.453 satırlık veri seti incelenmiştir.

### 1. Değişken Açıklamaları
* **Iron Feed:** Tesise giren ham cevherdeki demir oranı (Girdi).
* **Silica Feed:** Tesise giren ham cevherdeki silika oranı (Girdi).
* **Starch Flow & Amina Flow:** Ayrıştırma için kullanılan kimyasalların debisi.
* **Ore Pulp Flow/pH/Density:** Tank içindeki karışımın fiziksel özellikleri.
* **Flotation Column Air Flow:** Ayrıştırma tankına verilen hava miktarı.
* **Iron Concentrate:** Ürün içindeki demir oranı (Kalite göstergesi).
* **% Silica Concentrate (HEDEF):** Ürün içindeki safsızlık oranı. **Minimize edilmelidir.**

### 2. Dikkat Çeken Bulgular
* **Korelasyon Analizi:** Hedef değişkenimiz `% Silica Concentrate` ile `Iron Concentrate` arasında **güçlü negatif korelasyon** (-0.8 civarı) tespit edildi. Yani, demir konsantrasyonu arttıkça, silika (safsızlık) azalıyor. Bu, modelin en önemli özniteliğinin `Iron Concentrate` olacağının sinyalini veriyor.
* **Veri Dağılımı:** Hedef değişken (Silika) normal dağılıma yakın bir eğri çiziyor ancak zaman zaman %5'in üzerine çıkan aykırı değerler (kötü kalite anları) mevcut.
